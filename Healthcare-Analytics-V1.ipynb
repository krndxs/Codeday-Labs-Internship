```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Karandeep Singh\Downloads\archive (2)\healthcare\train_data.csv')
df.head()
df.dtypes
```

```python
df1 = df[['Severity of Illness', 'Type of Admission']]
df1.head()
df1.dropna()
```

```python
case_type = df1['Severity of Illness'].value_counts()
sns.set(style = 'whitegrid' )
sns.barplot(case_type.index, case_type.values)
plt.title('Frequency of Illness Status', fontsize = 16)
plt.ylabel('Frequency', fontsize=16)
plt.xlabel('Status', fontsize=16)
plt.show()
```

```python
case_type = df1['Type of Admission'].value_counts()
sns.set(style = 'whitegrid' )
sns.barplot(case_type.index, case_type.values)
plt.title('Frequency of Addmission Type', fontsize = 16)
plt.ylabel('Freq', fontsize=16)
plt.xlabel('Type', fontsize=16)
plt.show()
```

```python
df['Age'].value_counts(sort=False).plot.bar()
plt.title('Distribution of Age')
plt.ylabel('Count')
plt.xticks(fontsize = 12, rotation=0) 
plt.yticks(fontsize = 12)
```
